## Model Building 

```python
## calculate new column to represent time from signup to last visit date
from datetime import datetime 
store =[]
for x in range(len(data)):
    last_trip_date = datetime.strptime(data['last_trip_date'][x], '%Y-%m-%d') 
    signup_date = datetime.strptime(data['signup_date'][x], '%Y-%m-%d')
    store.append(last_trip_date - signup_date)
```

```python
store = pd.Series(store)
data['signup_to_last_date'] = store.dt.days
```

```python
data
```

```python
data.drop(columns = 'last_trip_date', inplace=True)
```

```python
data['city'].value_counts()
```

```python
# Convert categroical city data into dummy type variable. 
data = pd.get_dummies(data, columns=['city'])
```

```python
##Use on-hot encoding for phone and ultimate_black_user columns 
```

```python
data['phone'].replace({'iPhone':0,'Android':1}, inplace=True)
```

```python
data['ultimate_black_user'].replace({True:0, False:1}, inplace=True)
```

```python
data
```

```python
user_list = []
for x in data.signup_to_last_date:
    if x >=150 and x <=180:
        user_list.append(1)
    else:
        user_list.append(0)
```

```python
data['active_user'] = user_list
```

```python
data.active_user.value_counts() ## looks like there is a 3 to 1 ratio of inactive to active 
```

```python
data.drop(columns='signup_to_last_date', inplace=True)
```

```python
data.drop(columns='signup_date', inplace=True)
```

```python
data
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import classification_report
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
X = data.drop(columns='active_user')
y = data['active_user']
pipeline = make_pipeline(StandardScaler(), LogisticRegression(C=0.2, class_weight='balanced'))
y_pred = cross_val_predict(pipeline,X,y,cv=5)
print(classification_report(y,y_pred))
```

```python
from sklearn.preprocessing import RobustScaler
features = ['trips_in_first_30_days','avg_rating_of_driver','avg_surge','surge_pct','avg_dist','avg_rating_by_driver']
data_2 = data[features]
```

```python
transformer = RobustScaler(copy=False)
data_2 = transformer.fit_transform(data_2)
```

```python
data_2
```

```python
data_2 = pd.DataFrame(data_2, columns=features)
```

```python
data_2
```

```python
features_2 = ['phone','ultimate_black_user','city_Astapor',"city_King's Landing", 'city_Winterfell','weekday_pct','active_user']
addition = data[features_2]
```

```python
addition
```

```python
data_2 = data_2.join(addition)
```

```python
data_2
```

```python
data_2['active_user'].value_counts() ## this is an imbalanced dataset 
```

```python
X = data_2.drop(columns ='active_user')
y = data_2['active_user']
```

```python
clf = make_pipeline(StandardScaler(),LogisticRegression(class_weight='balanced', random_state=42))
scores = cross_val_score(clf,X,y,cv=5)
```

```python
print('The mean score for the model was %0.2f, with a std of %0.2f' % (scores.mean(),scores.std()))
```

```python
y_pred = cross_val_predict(clf,X,y,cv=5)
print(classification_report(y,y_pred))
```

```python
## I chose the logistic regression model because the model does not require the assumption of linearity nor does it require
## guassian errors. Logistic regression does well to not overfit the data. I considered a random forest method, but the 
## dataset was too small and could be subject to overfitting.  
```

```python
### The data for this model that displayed large number of outliers was scaled using the median and IQR range. Since the
### transformation resulted in negative intergers, we had to rescale the data to fit a standard normal distribution. We 
### could then fit and train using cross validation. The use of RobustScaler() yielded the same results as the model 
```

```python
coeff = abs(coef_model.named_steps['logisticregression'].coef_.flatten())
```

```python
feature_importance = dict(zip(names, coeff))
```

```python
feature_importance
```

```python
feature_importance = dict(sorted(feature_importance.items(), key=lambda item: item[1]))
```
